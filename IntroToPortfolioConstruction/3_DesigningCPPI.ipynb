# GBM Interactive plots

```python
erk.gbm(10, 100).plot(legend = False)
```

```python
def show_gbm(n_scenarios, mu, sigma):
    """
    Draw the results of a stock price evolution under GBM
    """
    s_0=100
    prices = erk.gbm(n_scenarios=n_scenarios, mu=mu, sigma = sigma, s_0 = s_0)
    ax = prices.plot(legend=False, color = 'indianred', alpha = .5, linewidth=2, figsize = (12,5))
    ax.axhline(y=s_0, ls=':', color='black')
    ax.plot(0, s_0, marker='o', color = 'darkred', alpha=.2)
```

```python
show_gbm(30, .07, .15)
```

```python
gbm_controls = widgets.interactive(show_gbm, 
                                  n_scenarios = (1, 20, 5),
                                  mu = (0, .2, .01),
                                  sigma = (0, .3, .01)
                                 )
```
